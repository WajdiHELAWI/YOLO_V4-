```python
import pandas as pd
import os 
```

```python
os.chdir('/content/drive/MyDrive/Yolo_v4/Dataset_and_Annotation/csv_folder')
classes_data = pd.read_csv('class-descriptions-boxable.csv', header=None)
classes_data.head()
```

```python
# Define required classes 
classes = ['Coffee', 'Tea']

#Get the string corresponding to the class name and store it inside a list called as class_srtings 
class_string = []
for i in classes:
    #Extracting the first column's elements for matching classes Airplane and Parachute
    req_classes = classes_data.loc[classes_data[1] == i]
    chaine = req_classes.iloc[0][0]
    print(chaine)
    class_string.append(chaine)

print(class_string)
```

```python
annotation_data = pd.read_csv('train-annotations-bbox.csv',usecols=['ImageID','LabelName','XMin','XMax','YMin','YMax'])
annotation_data.head()
```

```python
filtred_class_data = annotation_data.loc[annotation_data['LabelName'].isin(class_string)].copy()
filtred_class_data.head()
```

```python
filtred_class_data['classNumber'] = ''
filtred_class_data['center x'] = ''
filtred_class_data['center y'] = ''
filtred_class_data['width'] = ''
filtred_class_data['height'] = ''
```

```python
for i in range(len(class_string)):
  filtred_class_data.loc[filtred_class_data['LabelName'] == class_string[i], 'classNumber'] = i

filtred_class_data['centrer x'] = (filtred_class_data['XMax'] + filtred_class_data['XMin']) /2 
filtred_class_data['centrer y'] = (filtred_class_data['YMax'] + filtred_class_data['YMin']) /2 

filtred_class_data['width'] = (filtred_class_data['XMax'] - filtred_class_data['XMin'])  
filtred_class_data['height'] = (filtred_class_data['YMax'] - filtred_class_data['YMin'])

YOLO_values = filtred_class_data.loc[:,['ImageID','classNumber','centrer x','center y','width','height']].copy()
YOLO_values.head()
```

```python
images_path = '/content/drive/MyDrive/Yolo_v4/Dataset_and_Annotation/Dataset/train/Coffee_Tea'
os.chdir(images_path)
for current_dir ,dirs , files in os.walk('.'):
  for f in files : 
    if f.endswith('.jpg'):
      image_title = f[:-4]
      YOLO_file = YOLO_values.loc[YOLO_values['ImageID'] == image_title]
      df = YOLO_file.loc[:, ['classNumber','centrer x','center y','width','height']].copy()
      save_path = images_path + '/' + image_title +'.txt'
      df.to_csv(save_path,header=False,sep=' ')
```
